**Import Packages**

```python
import matplotlib.pyplot as plt
import numpy as np
```

**Example**

```python
#Data for plot 1
x1 = np.array([1, 2, 3, 4, 5])
y1 = np.array([4, 6, 10, 12, 18])

#Data for plot 2
x2 = np.array([1, 2, 3, 4, 5])
y2 = np.array([5, 4, 11, 13, 16])

#Data for plot 3
x3 = np.array([1, 2, 3, 4, 5])
y3 = np.array([10, 12, 14, 8, 7])
```

```python
#plot 1
plt.subplot(3, 1, 1)
plt.plot(x1,y1)
plt.title('plot 1')

#plot 1
plt.subplot(3, 1, 2)
plt.plot(x2,y2)
plt.title('plot 2')

#plot 1
plt.subplot(3, 1, 3)
plt.plot(x3,y3)
plt.title('plot 3')

plt.tight_layout()
plt.show()
```

```python
#plot 1
plt.subplot(1, 3, 1)
plt.plot(x1,y1)
plt.title('plot 1')

#plot 1
plt.subplot(1, 3, 2)
plt.plot(x2,y2)
plt.title('plot 2')

#plot 1
plt.subplot(1, 3, 3)
plt.plot(x3,y3)
plt.title('plot 3')

plt.tight_layout()
plt.show()
```
