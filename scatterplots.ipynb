**Import Packages**

```python
import matplotlib.pyplot as plt
import numpy as np
```

**Example 1**

```python
x = np.array([1, 2, 3, 4, 5])
y = np.array([4, 6, 10, 12, 18])
```

```python
plt.scatter(x,y)
plt.show()
```

```python
plt.scatter(x, y, color="green", marker = "s", edgecolors = "yellow")
plt.xlabel("X values")
plt.ylabel("Y values")
plt.show()
```
