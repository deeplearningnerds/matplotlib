## Import Packages

```python
import matplotlib.pyplot as plt
import numpy as np
```

## Defining Data

```python
labels = ["Python", "SQL", "Deep Learning", "Data Visualization"]
hours = [7, 6, 5, 2]
```

## A basic Pie Charts

```python
plt.pie(hours, labels=labels)
plt.show() 
```

## Highlight a specific category

```python
explode = (0.1, 0, 0, 0)

plt.pie(hours, explode=explode, labels=labels)
plt.show()
```

## Change colors, add the percentages of each  and add a legend

```python
colors = ["darkcyan", "aquamarine", "aqua", "skyblue"]

plt.pie(hours, explode=explode, colors=colors, labels=labels,
        autopct='%1.0f%%')
plt.legend(bbox_to_anchor=(1, 1), loc='upper left', borderaxespad=0)
plt.show()
```
