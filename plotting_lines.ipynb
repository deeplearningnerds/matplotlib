**Import Packages**

```python
import matplotlib.pyplot as plt
import numpy as np
```

**Example**

```python
x = np.array([1, 2, 3, 4, 5])
y = np.array([4, 6, 10, 12, 18])
```

```python
plt.plot(x,y)
```

```python
plt.plot(x, y, color="red", linestyle="--")
plt.xlabel("X values")
plt.ylabel("Y values")
plt.show()
```

```python
plt.plot(x, y, marker="s",color="green")
plt.xlabel("X values")
plt.ylabel("Y values")
plt.show()
```

```python
plt.plot(x, y,"o")
plt.xlabel("X values")
plt.ylabel("Y values")
plt.show()
```

```python
plt.plot(x, y, marker="h",color="red", markerfacecolor="yellow", markeredgecolor='black')
plt.xlabel("X values")
plt.ylabel("Y values")
plt.show()
```
