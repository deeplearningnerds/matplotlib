## Import Packages

```python
import matplotlib.pyplot as plt
```

## Example 1

```python
labels = ["2018", "2019", "2020", "2021"]
users_tool = [110, 100, 240, 310]
```

```python
plt.bar(labels, users_tool, color='orange')
plt.xlabel("Years")
plt.ylabel("Users")
plt.title('Users by years')
plt.show()
```
